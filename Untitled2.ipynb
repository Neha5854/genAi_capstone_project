```python
import pandas as pd

try:
    # Assuming the file is in the root of your Google Drive
    df = pd.read_csv('train.csv')
    print("train.csv loaded successfully.")
    print(df.head())
except FileNotFoundError:
    print("Error: train.csv not found. Please make sure the file is in the correct directory in your Google Drive.")
    df = None # Ensure df is None if file not found
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

if df is not None:
    # Handle potential missing values in 'catalog content'
    df['catalog_content'] = df['catalog_content'].fillna('')

    tfidf_vectorizer = TfidfVectorizer(max_features=5000) # Limit features to manage memory
    tfidf_matrix = tfidf_vectorizer.fit_transform(df['catalog_content'])
    print("\nTF-IDF vectorization complete.")
    print(f"Shape of TF-IDF matrix: {tfidf_matrix.shape}")
else:
    print("\nSkipping TF-IDF vectorization as data was not loaded.")
    tfidf_matrix = None
```

```python
from sklearn.cluster import KMeans

if tfidf_matrix is not None:
    # Choosing a reasonable number of clusters (can be adjusted)
    n_clusters = 5
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10) # Set n_init explicitly
    kmeans.fit(tfidf_matrix)
    print(f"\nK-means clustering complete with {n_clusters} clusters.")
    df['cluster'] = kmeans.labels_
    print("\nFirst 5 rows with cluster labels:")
    print(df[['catalog_content', 'cluster']].head())
else:
    print("\nSkipping K-means clustering as TF-IDF matrix is not available.")
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans # Import KMeans to perform clustering for different n_clusters

if tfidf_matrix is not None:
    # Define the range of cluster numbers to visualize
    cluster_numbers_to_visualize = range(2, 10) # Corresponds to 2 through 9

    # Reduce dimensionality using PCA (fit on the full matrix for consistency)
    pca = PCA(n_components=2)
    pca_components = pca.fit_transform(tfidf_matrix.toarray())

    for n_clusters in cluster_numbers_to_visualize:
        print(f"\nPerforming K-means clustering with {n_clusters} clusters for visualization.")
        # Perform KMeans clustering for the current number of clusters
        kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
        kmeans.fit(tfidf_matrix)
        cluster_labels = kmeans.labels_

        # Create a scatter plot of the clusters
        plt.figure(figsize=(10, 8))
        plt.scatter(pca_components[:, 0], pca_components[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.6)
        plt.title(f'K-means Clustering of Catalog Content ({n_clusters} Clusters, PCA Reduced)')
        plt.xlabel('PCA Component 1')
        plt.ylabel('PCA Component 2')
        plt.colorbar(label='Cluster Label')
        plt.show()
else:
    print("\nSkipping visualization as TF-IDF matrix is not available.")
```

```python
from sklearn.metrics import silhouette_score

if tfidf_matrix is not None and 'cluster' in df.columns:
    # Calculate silhouette score
    try:
        silhouette_avg = silhouette_score(tfidf_matrix, df['cluster'])
        print(f"\nSilhouette Score: {silhouette_avg}")
    except Exception as e:
        print(f"\nError calculating silhouette score: {e}")
        print("This can happen if there is only one cluster or less than 2 samples.")
else:
    print("\nSkipping silhouette score calculation as TF-IDF matrix or cluster labels are not available.")
```

```python
from yellowbrick.cluster import SilhouetteVisualizer
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import time # Import time for adding pauses
from sklearn.metrics import silhouette_samples # Import silhouette_samples

if tfidf_matrix is not None:
    # Define a range of cluster numbers to evaluate
    range_n_clusters = [2, 3, 4, 5, 6, 7, 8, 9]

    # Reduce the size of the tfidf_matrix for faster computation
    # Using a smaller sample size for silhouette analysis
    sample_size = 10000 # You can adjust this number
    if tfidf_matrix.shape[0] > sample_size:
        # Randomly sample indices
        np.random.seed(42) # for reproducibility
        sample_indices = np.random.choice(tfidf_matrix.shape[0], size=sample_size, replace=False)
        tfidf_matrix_sample = tfidf_matrix[sample_indices]
    else:
        tfidf_matrix_sample = tfidf_matrix


    # Iterate over the range of cluster numbers
    for i in range_n_clusters:
        print(f"Generating silhouette plot for {i} clusters...")
        # Create KMeans instance for the current number of clusters
        km = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=100, random_state=42)

        # Fit the KMeans model
        cluster_labels = km.fit_predict(tfidf_matrix_sample)

        # Calculate silhouette scores for each sample
        sample_silhouette_values = silhouette_samples(tfidf_matrix_sample, cluster_labels)

        # Print average silhouette score for each cluster
        print(f"Average Silhouette Scores for {i} clusters:")
        for cluster_label in range(i):
            # Aggregate the silhouette scores for samples belonging to this cluster
            cluster_silhouette_values = sample_silhouette_values[cluster_labels == cluster_label]
            average_cluster_silhouette_score = cluster_silhouette_values.mean()
            print(f"  Cluster {cluster_label}: {average_cluster_silhouette_score:.4f}")


        # Create a figure and axes for the current subplot
        fig, ax = plt.subplots(1, 1, figsize=(8, 6)) # Create a single subplot for each plot

        # Create SilhouetteVisualizer instance with KMeans instance and subplot position
        visualizer = SilhouetteVisualizer(km, colors='yellowbrick', ax=ax)

        # Fit the visualizer on the sampled data (re-fitting the visualizer after km.fit_predict)
        visualizer.fit(tfidf_matrix_sample, cluster_labels) # Pass labels to avoid re-fitting km

        # Set the title for the subplot
        visualizer.show()

        # Close the figure to free up memory
        plt.close(fig)

        # Add a small pause
        time.sleep(1) # Pause for 1 second

    print("Finished generating silhouette plots and printing scores.")
else:
    print("TF-IDF matrix is not available. Skipping silhouette plot and score calculation.")
```
